```python
%matplotlib inline
from numpy.random import gumbel, normal
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(rc={'figure.facecolor':'#F2F2F2'})

gumbel_strengths = np.array([gumbel(-1, size=10000), gumbel(1, size=10000)]).T
sns.distplot(gumbel_strengths[:,0], label='Gumbel Strength 1')
sns.distplot(gumbel_strengths[:,1], label='Gumbel Strength 2')

normal_strengths = np.array([normal(-1, size=10000), normal(1, size=10000)]).T
sns.distplot(normal_strengths[:,0], label='Normal Strength 1')
sns.distplot(normal_strengths[:,1], label='Normal Strength 2')
plt.legend()
```

```python
normal_differences = normal_strengths[:,0] - normal_strengths[:,1]
gumbel_differences = gumbel_strengths[:,0] - gumbel_strengths[:,1]

sns.distplot(normal_differences, label='Normal Differences')
ax = sns.distplot(gumbel_differences, label='Gumbel Differences')
ax.legend()
```

```python
normal_hist, normal_bin_edges = np.histogram(normal_differences, bins=100)
gumbel_hist, gumbel_bin_edges = np.histogram(gumbel_differences, bins=100)
sns.lineplot(x=normal_bin_edges[1:], y=np.cumsum(normal_hist)/np.max(np.cumsum(normal_hist)), label='Normal CDF')
sns.lineplot(x=gumbel_bin_edges[1:], y=np.cumsum(gumbel_hist)/np.max(np.cumsum(gumbel_hist)), label='Logistic CDF')
```
